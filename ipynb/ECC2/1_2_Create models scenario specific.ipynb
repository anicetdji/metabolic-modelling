```python
name_model = 'PKT_o2lim_ac_ko'
souche = name_model.split('_')[0]
model = cobra.io.read_sbml_model(wd+'files/models/model_'+souche+'_Isopropanol_iML1515.xml')

if name_model.count('o2lim')>0:
    model.reactions.EX_o2_e.bounds=(-10,0)
if name_model.count('_ac')==0:
    model.reactions.EX_ac_e.bounds=(-10,0)
    model.reactions.EX_acac_e.bounds=(-10,0)
if name_model.count('dels')>0:
    model.reactions.LDH_D.bounds = (0,0)
    model.reactions.ALCD2x.bounds = (0,0)
    model.reactions.FRD2.bounds = (0,0)
    model.reactions.FRD3.bounds = (0,0)
if name_model.count('PFL')>0:
    model.reactions.PFL.bounds = (0,0)
if name_model.count('iprop50')>0:
    model.reactions.LDH_D.bounds = (0,0)
    model.reactions.ALCD2x.bounds = (0,0)
    model.reactions.PFL.bounds = (0,0)
if name_model.count('_ko')>0:
    model.reactions.PFL.bounds = (0,0)
    model.reactions.ACKr.bounds = (0,0)
    model.reactions.LDH_D.bounds = (0,0)
    model.reactions.ALCD2x.bounds = (0,0)
    model.reactions.PDH.bounds = (80,1000)
```

```python
# try pfba
res = cameo.pfba(model)
# Save Model
cobra.io.save_json_model(model, wd+'input/models/model_'+name_model+'.json')
```
