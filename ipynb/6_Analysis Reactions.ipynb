```python
model = cameo.load_model('iML1515')
```

```python
df_exc = pd.DataFrame(index=[r.id for r in model.reactions if r.id.startswith('EX_')])
for r_id in df_exc.index:
    m = list(model.reactions.get_by_id(r_id).metabolites.keys())[0]
    df_exc.loc[r_id,'m_id'] = m.id.split('_')[0]
    df_exc.loc[r_id,'m_name'] = m.name
        
print(len(df_exc))
df_exc.head()
```

```python
df_exc.reindex(x for x in df_exc.index if df_exc.loc[x,'m_name'].lower().count('folate')>0)
```
